```python
df_distr = pd.DataFrame(columns=lst_disaster, index=lst_source)
```

```python
dir_news = '/home/admin/sampling-data/output/output_20200414'
for disaster in lst_disaster:
    for source in lst_source:
        file_name = source + ".json" + "_" + disaster + ".txt"
        n_lines = sum(1 for line in open(dir_news + "/" + file_name))
        df_distr.at[source, disaster] = n_lines

df_distr
```

```python
lst_tmp = []
boundary = 20
df = df_distr
```

```python
for column in lst_disaster:
    lst_tmp = []
    for index, row in df.iterrows():
        val = row[column]
        if (val < boundary):
            lst_tmp.append(val)
        else:
            lst_tmp.append(boundary)

    df_sample = pd.DataFrame(lst_tmp, columns=['sample_' + column], index=lst_source)
    df['sample_' + column] = df_sample
```
